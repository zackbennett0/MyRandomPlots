# Random Scatter Plot Code

Hello, this is my Random Plot code, where a scatter plot with new values will appear each
time the code is run. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Create data
N = 20
x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area)
plt.title('Scatter plot of a data set')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
